```python
### Step by step guide for reference 

import pandas as pd

df =  pd.read_csv('pokemon_data.csv')


# pd.read_excel for xlsx format
# df = pd.read_csv('pokemon_data.txt,delimiter = '\t')
# print(df)
# print(df.head(3))

```

```python
## To Read Columns
df.columns
print(df.columns)
print('---------------')
print(df['Type 1'][0:6])
print('---------------')
print(df[['Type 1','Type 2']][0:10])
print('---------------')
print(df[['Name','Attack','Defense']])
print('---------------')
#print(df.Name)  for one words
```

```python
## To Read Rows

#read dataset
print(df.head(4))
print('---------------')

#access position
print(df.iloc[30])
print('---------------')

#access range of position
print(df.iloc[99:103])
print('---------------')

#access position and column
print(df.iloc[99,2])
print('---------------')

# print index and name row in for loop
for index, row in df.iterrows():
    print(index,row['Name'])

print('---------------')

# print by criterion
print(df.loc [df['Type 1'] == 'Bug'])
print('---------------')
print(df.loc [df.Legendary == True])
```

```python
## Sort & Describe Data

# df.describe()

# df.sort_values('Name', ascending = False)

# sort by 2 variables
df.sort_values(['Type 1', 'Attack'], ascending= [1,0])
```

```python
## Making Changes to Data
# added the total stats

df['Base Total'] = df['HP'] + df['Attack'] + df['Defense']
df.head(5)

df = df.drop (columns = ['Base Total'])
# : is all rows
df['Base Total'] = df.iloc[:, 4:10].sum(axis=1)
#print(df['Base Total'])

#Rearrange columns
cols = df.columns.values
df = df[cols[0:4] + [cols[-1]]+cols[4:12]]
print(df)
```

```python
## Filtering Data & , | , and or
newdf = df.loc[( df['Type 1'] == 'Fire') |(df['Type 2'] == 'Dragon') ]

print(newdf)
## Reset Index

newdf = newdf.reset_index()
print(newdf)

## search by string
df.loc[df["Name"].str.contains("chu")]

## not ~
df.loc[~df["Name"].str.contains('Mega')]

## regex is used for search patterns
## flags = re.I is used for ignore capitals
import re
df.loc[df['Type 1'].str.contains('stEEl|Ghost', flags = re.I, regex = True)]

## ^ for starts with
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
print(df)
```

```python
## Conditional Changes
# rename grass to salad
df.loc[df['Type 1'] == 'Grass', 'Type 1'] = 'Salad' 

# change condition legendary
df.loc[df['Type 1'] == 'Salad', 'Legendary'] = True
df.head(5)

# change multiple value
df.loc[df['Attack'] < 50, ['Total', 'Base Total']] = ['Weak', 'Nevermind']

df.head(10)
```

```python
## Aggregate Statistics Groupby
# MEAN
df.groupby(['Type 1']).mean().sort_values('Attack',ascending = False)


# SUM
df.groupby(['Type 1']).sum()

# Count
df['count'] =1
df.groupby(['Generation', 'Type 1']).count()['count']
```
